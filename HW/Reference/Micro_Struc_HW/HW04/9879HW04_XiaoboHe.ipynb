## Glosten and Milgrom

### 1. (8 points) 

(Adapted from Hasbrouck 5.2 but please note that the setup in (3) in the slides is different from Hasbrouck's in Figure 5.1 of his book.) Consider a variant of the Glosten and Milgrom model where there is informed trading only in the low state $V=\underline V$.

(a) Show that the learned probability after one step that the true price is $\bar V$ is given by

$$
\delta_1(Buy)=\Pr(\bar V|Buy)=\frac{\delta}{1-\mu\,(1-\delta)};\quad \delta_1(Sell)=\Pr(\bar V|Sell)=\frac{\delta}{1+\mu\,(1-\delta)}
$$

and that

$$
A=\frac{\underline V\,{(1-\delta)\,(1-\mu)}+\bar V\,{\delta}}{{1-\mu\,(1-\delta)}};\quad B=\frac{\underline V\,{(1-\delta)\,(1+\mu)}+\bar V\,{\delta}}{1+\mu\,(1-\delta)}
$$

(b) Deduce an expression for the spread $s=A-B$.

(c) Explain the sensitivities of the spread to the parameters of the model to a non-expert.


### 1. Solution
### (a)

$$\delta_1(Buy)=\Pr(\bar V|\text{Buy})
=\frac{\Pr(\bar V \cap \text{Buy})}{\Pr(\text{Buy})}
=\frac{\frac 12 \delta}{\frac 12[1-\mu(1-\delta)]}
=\frac{\delta}{1-\mu\,(1-\delta)}$$



$$\delta_1(Sell)=\Pr(\bar V|\text{Sell})
=\frac{\Pr(\bar V \cap \text{Sell})}{\Pr(\text{Sell})}
=\frac{\frac 12 \delta}{\frac 12 [1+\mu(1-\delta)]} 
=\frac{\delta}{1+\mu\,(1-\delta)}$$

$$\Pr(\underline V|Buy)=1-\Pr(\bar V|\text{Buy})=\frac{(1-\delta)(1-\mu)}{1-\mu(1-\delta)}$$

$$\Pr(\underline V|Sell)=1-\Pr(\bar V|\text{Sell})=\frac{(1-\delta)(1+\mu)}{1-\mu(1-\delta)}$$

$$A = \mathbb{E} [V|Buy] = {\bar V\,\Pr(\bar V|Buy)+\underline V\,\Pr(\underline V|Buy)} = \frac{\underline V(1-\delta)(1-\mu)+\bar V\delta}{1-\mu(1-\delta)} $$

$$B= \mathbb{E} [V|Sell] = {\bar V\,\Pr(\bar V|Sell)+\underline V\,\Pr(\underline V|Sell)}= \frac{\underline V\,{(1-\delta)\,(1+\mu)}+\bar V\,{\delta}}{1+\mu\,(1-\delta)}$$

### (b)

The spread:
$$ s = A-B =  \frac{\underline V(1-\delta)(1-\mu)+\bar V\delta}{1-\mu(1-\delta)} - \frac{\underline V\,{(1-\delta)\,(1+\mu)}+\bar V\,{\delta}}{1+\mu\,(1-\delta)} = \frac{2\delta\mu(1-\delta)(\bar V - \underline V)}{1-\mu^2(\delta-1)^2} $$

### (c)
The spread will increase as $(\bar V - \underline V)$ increases, as a result to compensate the higher risk associated with larger difference between $\bar V$ and $\underline V$.

Spread will decrease as $\delta$ increases which means there is more "adverse selection" in the $\underline V$ stage.

Spread will increase as $\mu$ increase because more informed trading in the market and market maker needs larger compensation with bigger spread.

## The Lee and Ready tick test

### 2. (10 points) 

In the following use the trades and quotes object *tqBAC* in the loaded data *tq.rData*.  Denote trade prices by $p_t$ and mid-quotes by $m_t$.

(a) Compute the time series of trade signs $\epsilon_t$ according to $\epsilon_t=+1$ if $p_t>m_t$ and  $\epsilon_t=-1$ otherwise, using the Lee-Ready rule for trades at mid.  Name this series *tradeSigns*. (Note this was already done in the lecture.)

(b) Compute the Lee-Ready trade-sign indicator time series (as if you didn't have quotes and so no mid-prices).

(c) Compute the correlation of these two indicators.

(d) Assume the rule in part (a) is 100% accurate.  What percentage of trades are correctly signed by the Lee-Ready test?

(e) Explain the relationship between your answers in (c) and (d).


### 2. Solution

```r
#(a)
#Time series of trade signs
tradeSigns <- getTradeDirection(tqBAC) 
```

```r
#(b)
#Lee-Ready trade-sign indicator time series
price <- as.numeric(tqBAC$PRICE)
len_price <- length(price)
indicator <- rep(0,length(price))
indicator[1] <- 1

for(i in 2:len_price){
    if (price[i] > price[i-1])
        indicator[i] <- 1
    else if (price[i] < price[i-1])
        indicator[i] <- -1
    else
        indicator[i] <- indicator[i-1]   
}
```

```r
#(c)
#Correlation of these two indicators
corr <- cor(indicator, tradeSigns)
sprintf("Correlation of these two indicators is: %s", corr)
```

```r
#(d)
#Percentage of trades are correctly signed by the Lee-Ready test
counter_diff <- 0

for (i in 1:len_price){
    if (indicator[i] != tradeSigns[i])
        counter_diff <- counter_diff + 1
}
    
percent <- 1-counter_diff/len_price
sprintf ("Percentage of trades are correctly signed by the Lee-Ready test: %s", percent)
```

### (e)
When the percentage of trades correctly signed are higher, the correlation is higher. It is cosistent with (c) and(d) since the correlation is not 1.

```r
#(a)
# Generate rsLag
rsLag <- function(exch) {
    set <- subset(tqBAC, tqBAC$EX==exch)
    mids <- (as.numeric(set$OFR) + as.numeric(set$BID))/2
    price <- as.numeric(set$PRICE)
    len_price <- length(price)
    tradeSigns <- getTradeDirection(set)
    rs <- function(lag){mean(tradeSigns[1:(len_price-lag)]*(price[1:(len_price-lag)]-mids[-(1:lag)]))}
    rsLag <- sapply(1:50,rs)
    return (rsLag)
}

# create rsLags for different exchanges
rsLag_D <- rsLag('D')
rsLag_B <- rsLag('B')
rsLag_Z <- rsLag('Z')
rsLag_Y <- rsLag('Y')
rsLag_P <- rsLag('P')

#Plot the figures
plot(x=1:50,y=rsLag_D,col="red",type="l",xlab=expression(tau), ylab = "Realized spread", ylim=c(-0.003, 0.008))
lines(x=1:50,y=rsLag_B,col="blue")
lines(x=1:50,y=rsLag_Z,col="orange")
lines(x=1:50,y=rsLag_Y,col="purple")
lines(x=1:50,y=rsLag_P,col="green")
legend("topleft", 
       c("D-FINRA",
         "B-NASDAQ OMX BX",
         "Z-BATS Exchange",
         "Y-BATS Y-Exchange",
         "P-NYSE Arca SM"),
       lty = c(1,1,1,1,1), 
       col=c("red","blue","yellow","purple","green"))
```

### 4. Solution
### (a) 
The conditional expectation $\E[\epsilon_t|\epsilon_{t-1}]$ in terms of $\pi$:
$$\mathbb{E} [\epsilon_t|\epsilon_{t-1}] = \epsilon_{t-1}(1-\pi) - \epsilon_{t-1}\pi = \epsilon_{t-1}(1-2\pi) $$


```r
#(b)
tradeSigns <- getTradeDirection(tqBAC) 
epsilon_t <- tradeSigns[1:(length(tradeSigns)-1)]
epsilon_t_1 <- tradeSigns[2:length(tradeSigns)]

df_tradeSigns <- data.frame(epsilon_t, epsilon_t_1)
df_tradeSigns.lm <- lm(epsilon_t ~ epsilon_t_1, data=df_tradeSigns)
pi_empirical <- (1 - coefficients(df_tradeSigns.lm)[2])/2.0

sprintf("Coefficient: %s", coefficients(df_tradeSigns.lm)[2])
sprintf("Empirical pi: %s",pi_empirical)
```
