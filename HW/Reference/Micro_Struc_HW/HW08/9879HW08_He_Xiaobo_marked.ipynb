## Predictability of order flow

### 1. (6 points) <font color=blue> Score: 6/6 </font>

Suppose that the distribution of meta order sizes $Q$ is exponential so that

$$
\Pr(Q>V)= \frac{1}{\lambda}\,e^{-\lambda\,V}
$$

Assuming slices of  constant size $\bar v$, compute the conditional probability that a hidden order has more slices given that $n$ slices have already been observed.  Compare your result with the case where distribution of metaorder size is power-law.

### 1. Solution:
If the distribution of meta order sizes is exponential, we have 

$$
\begin{eqnarray}
\mathbb P(Q \geq (n+1) \bar{v} | Q \geq n\bar{v}) 
= \frac{ \mathbb P( Q \geq (n+1) \bar{v})}{\mathbb P(Q \geq n\bar{v})}
= \frac{ \frac{1}{\lambda} e^{-\lambda(n+1) \bar{v}}}  {\frac{1}{\lambda} e^{ -\lambda n\bar{v}}}
= e^{-\lambda \bar{v}}
\end{eqnarray}
$$

If the distribution of meta order sizes is power-law, we have

$$
\begin{eqnarray}
\mathbb P(Q \geq(n+1) \bar{v} | Q \geq n\bar{v}) 
= \frac{\mathbb P(Q \geq(n+1) \bar{v})}{\mathbb P(Q \geq n\bar{v})}
\sim (\frac {n} {n+1}) ^ {\beta}
\end{eqnarray}
$$

Summary: in the exponential distribution case, the probability of getting one more slice after $n$ slices is independent of the previous $n$ slices; while in power-law distribution case, as $n$ increases, the probability of getting one more slice after $n$ slices is close to 1.

## The Obizhaeva-Wang model

### 3. (8 points) <font color=blue> Score: 8/8 </font>

In the Obizhaeva Wang (OW) model, the price follows

$$
S_t=S_0+\eta\,\int_0^t\,{\dot x}_s\,e^{-\rho\,(t-s)}\,ds+\int_0^t\,\sigma\,dZ_s
$$

(a) Derive an expression for the expected price path corresponding to a VWAP execution.

(b) With the duration $T$ of the meta order set to 1, plot the expected price path from time $t=0$ to time $t=2$ for $\rho=0.5,1,$ and $2$.

(c) What is the expected average cost per share $\bar S_T$ of a VWAP order in the OW model?

### 3. Solution
### (a)
Under VWAP, $\dot x_s = v$, where v is the constant excution speed.
$${\dot x}_s =\left \{ \begin{array}{rcl}
 \frac{n}{T} & \mbox{for} & s\leq T \\ 
0 & \mbox{for} & s>T 
\end{array}\right.  $$

Therefore, when $ t \leq T$,
$$
\begin{eqnarray}
\mathbb E [S_t] 
&=& S_0 +  \mathbb E [\eta \int_0^t \dot x_s e^{-\rho (t-s)}ds]
= S_0 + \eta v \int_0^t e^{-\rho(t-s)} ds 
= S_0 + \frac {\eta v} {\rho} (1-e^{-\rho t}).
\end{eqnarray}
$$

While when $ t \geq T$,
$$
\begin{eqnarray}
\mathbb E [S_t] 
&=& S_0 +  \mathbb E [\eta \int_0^t \dot x_s e^{-\rho (t-s)}ds]
= S_0 + \eta v \int_0^t e^{-\rho(t-s)} ds 
= S_0 + \frac {\eta v} {\rho} e^{-\rho t}(e^{\rho T} -1).
\end{eqnarray}
$$



```r
### (b)
#expected price move
price_path <- function(t,rho){
    if ((t >= 0) && (t <= 1)){ 
        px <- (1-exp(-rho*t)) / rho;
    }
    else {
        px <- exp(rho*(1-t))/rho - exp(rho*(-t))/rho; 
    }
    return (px)
}

times_step <- seq(0,2,0.001)
rho1 <- 0.5
rho2 <- 1.0
rho3 <- 2.0

s1 <- sapply(times_step,  function(t){price_path(t, rho1)})
s2 <- sapply(times_step,  function(t){price_path(t, rho2)})
s3 <- sapply(times_step,  function(t){price_path(t, rho3)})

plot(times_step, s1, type='l', xlab='time', ylab='expected price path', col='yellow')
points(times_step, s2, type='l', col='red')
points(times_step, s3, type='l', col='blue')
legend("topleft", c("rho=0.5","rho=1.0","rho=2.0"),col=c("yellow","red","blue"),lty =1)
```
